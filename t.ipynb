```python
def matrixpower(A,n):
    #b = A ..this is a shallow copy
    B =[i[:]for i in A]
    #c=a
    C = [i[:]for i in A]
    for i in range(n-1):
        C[0][0] = A[0][0]* B[0][0] + A[0][1]* B[1][0]
        C[0][1] = A[0][0]* B[0][0] + A[0][1]* B[1][1]
        C[1][0] = A[1][0]* B[0][0] + A[1][1]* B[1][0]
        C[1][1] = A[1][0]* B[0][1] + A[1][1]* B[1][1]

        B = [i[:] for i in C]


    return C 
    
A = [[1,1],[1,0]]

print(matrixpower(A,1))
print(matrixpower(A,2))
print(matrixpower(A,3))

```

```python
#
for n in range(10):
    result = matrixpower(A,n)
    print(n,result)
```
