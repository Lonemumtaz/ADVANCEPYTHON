```python
# CONTROL FLOW :  DETERMINES THE ORDER OF EXECUTION IN WHICH STATEMENTS ARE EXECUTED
# 1. if statement
# 2. if-else
# 3. if elif else
# 4. nested if
```

```python

# if Statement
number = 13
guess = int (input("guess number:"))

if guess == number:
    print("congo!!!")
    print("perfect")


print("Done")    
```

```python

# if-else Statement :used when we have two posssible outcomes

number = int (input("enter number:"))

if number%2 == 0:
    print("Even!")
else:
    print("ODD!")
```

```python

#if elif else : used when we have multiple conditions

marks = int(input("enter marks out of 100:"))

if marks > 90:
    print("Grade : 0")
elif marks > 70:
    print("Grade : A")
elif marks > 60:
    print("Grade : B")    
else:
    print("Fail")        
```

```python
# Nested if :if statement inside if 

age = 18
citizen = True

if age > 17:
    if citizen == True:
       print("Eligible!")
```
