2. prompt the user :what is your preferred unit of  height? Type "F" for feet and "M" for meters andd record the preference
3. if user  syas "F" then prompt user "you will enter height given as feet and inches. First enter feet and record feet and than prompt the user "Now enter the inches " and record inches 
4. if user says "M" then prompt user " what is your height in meters" and record in meters.
5. if user had choosen "F" than convert height in meters.
6. compute BMI using the following formulla

```python
weight = float(input("Enter the weight in kgs:"))
print(weight)
```

```python
preferredHeight = input("prefered unit of height is:")
print(preferredHeight)
```

```python
if preferredHeight == "F":
    feet , inch  = input("Enter height in feet and inches.").split()
    feet = float(feet)
    inch = float(inch)
    inches = (feet*12) + inch
    centimeters = inches* 2.5
    height= centimeters/100
    print(height)
    
elif preferredHeight == "M":
    height= input("Enter height in meters:")
    height = print(float(height))
    print(height)
```

```python
BMI = weight/(height*height)
print(BMI)
```

```python
if BMI <18.5:
    print("UNDERWEIGHT")
elif BMI < 25:
    print("NORMAL")
elif BMI < 30:
    print("OVERWEIGHT")
elif BMI > 30:
    print("VERY OVERWEIGHT")            
```
