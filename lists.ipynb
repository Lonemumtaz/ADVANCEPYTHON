***list inside lists***


```python
#CREATING A 2-D LIST
gatecs = ['ada',['clr', 'ss'], 'DM' , [ 'RS','sh']]
print(gatecs)
```

```python
# working with 2-D lists
#using for loop and  print individuals elements without printing the lists  
for i in gatecs:
    if isinstance(i,list):
     for j in i:
      print(j)
    else:
       print(i)
```

```python
#creating a 3-D list


gate = ['ada',['clr','ss',[12]] ,'DM',['RS','SH'],[15]]
print(gate)
```

```python
#working with 3-D lists using loop and  print individuals elements without printing the lists
for i in gatecs:
    if isinstance(i,list):
     for j in i:
          if isinstance(j,list):
             for k in j:
                 print(k)
          else:
               print(j)  
    else:
       print(i)
```

```python
# working with 3-D lists using  fumction to print the elements of the list without printing the lists

def printlist(gate):
    for i in gate:
     if isinstance(i,list):
      for j in i:
          if isinstance(j,list):
             for k in j:
                 print(k)
          else:
               print(j)  
    else:
       print(i)

printlist(gate)
```

```python
# using recurrsion to print the elements of the list without printing the lists
#recursion is a function that calls itself until it reaches a base case
#MAKE A RECURSIVE FUNCTION TO PRINT THE ELEMENTS OF THE LIST WITHOUT PRINTING THE LISTS
def printlist(gate):
    for i in gate:
     if isinstance(i,list):
      printlist(i)
     else:
      print(i)

printlist(gate)     
```
